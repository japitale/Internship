```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from time import sleep
from random import randint
from requests import get
```

```python
pos = []
player = []
team = []
rating = []
#data = []
```

```python
url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi'
html_content = requests.get(url).text
soup = BeautifulSoup(html_content,'lxml')
print(soup.prettify())
```

```python
#toper = soup.find('div',attrs={"class":"rankings-block__banner--player-info"})
toper = soup.find('div',attrs={"data-cricket-role":"batting"})

name = toper.find(class_='rankings-block__banner--name').text
#temp_name1 = toper.find(class_='rankings-block__banner--name').text
player.append(name)

nation = toper.find(class_='flag-15')
#temp_nation1 = toper.find(class_='flag-15')
team.append(nation)

top_rating = toper.find(class_='rankings-block__banner--rating').text
#temp_rate1=toper.find(class_='rankings-block__banner--rating').text
rating.append(top_rating)
```

```python
#temp_rate1
```

```python
player_table = toper.find('table',attrs={"class":"table"})
for x in player_table:
    rows = player_table.find_all('tr')[1:11]
    data = []
    for tr in rows:
        #data=[]
        cols = tr.find_all('td')
        for td in cols:
            #td.replace('\n$', '')
            data.append(td.text.strip())
            #pos.append(data[0])
          
```

```python
data
```

```python
#pos = data[0::4]
#while data:
temp_name2=[]
temp_nation2=[]
temp_rate2=[]

temp_name2 = (data[1::4])
temp_nation2 = (data[2::4])
temp_rate2 = (data[3::4])

player.extend(temp_name2)
team.extend(temp_nation2)
rating.extend(temp_rate2)
```

```python
#rating
```

```python
player_data = pd.DataFrame({#'Position': pos,
                       'Player': player,
                       'Team': team,
                       'Rating': rating
                       })

player_data
```
