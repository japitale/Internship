```python
import selenium
import pandas as pd
from selenium import webdriver 
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(r"C://chromedriver.exe")
driver.maximize_window()
driver.get('https://github.com/')
```

```python
trend = driver.find_element_by_xpath("//ul[@class = 'list-style-none mb-3'][2]//li[3]//a")
trending = trend.get_attribute('href')
driver.get(trending)
```

```python
Repository_title = []
Repository_description = []
Contributors_count = []
Language_used = []
```

```python
title = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")

try:
    for i in title:
        repo_title = i.text
        Repository_title.append(repo_title)

except NoSuchElementException as e:
        Repository_title.append("-")
  
```

```python
try:
    desc = driver.find_elements_by_xpath("//p[@class = 'col-9 color-text-secondary my-1 pr-4']")
    for i in desc:
        repo_desc = i.text
        Repository_description.append(repo_desc)
        
except NoSuchElementException as e:
        Repository_description.append("-")
```

```python
try:
    lang = driver.find_elements_by_xpath("//div[@class = 'f6 color-text-secondary mt-2']/span/span[2]")
    for i in lang:
        language = i.text
        Language_used.append(language)
        
except NoSuchElementException as e:
        Language_used.append("-")
```

```python
github= pd.DataFrame({})
github['Repository_title']= Repository_title[0:20]
github['Repository_description']= Repository_description[0:20]
github['Language_used']= Language_used[0:20]


github
```
