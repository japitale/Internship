```python
import selenium
import pandas as pd
from selenium import webdriver 
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(r"C://chromedriver.exe")
driver.get('https://www.naukri.com/')
```

```python
rec = driver.find_element_by_xpath("//a[@title = 'Browse All Recruiters']")
driver.get(rec.get_attribute('href'))
```

```python
skills = driver.find_element_by_xpath("//input[@class = 'sugInp']")
skills.clear()
skills.send_keys("Data Science")

search = driver.find_element_by_id('qsbFormBtn')
search.click()
```

```python
Rec_Name = []
Designation =[]
Company =[]
Skills =[]
Location = []
```

```python
def data():
    
    try:
        rname = driver.find_elements_by_xpath("//span[@class = 'fl ellipsis']")

        for i in rname:
            r_name = i.text
            Rec_Name.append(r_name)
    
    except NoSuchElementException as e:
        Rec_Name.append("-")
        
        
    
    try:
        des = driver.find_elements_by_xpath("//span[@class = 'ellipsis clr']")

        for i in des:
            desg = i.text
            Designation.append(desg)
    
    except NoSuchElementException as e:
        Designation.append("-")
    
    
    try:
        comp = driver.find_elements_by_xpath("//p[@class = 'highlightable']//a[2]")

        for i in comp:
            comp_name = i.text
            Company.append(comp_name)
    
    except NoSuchElementException as e:
        Company.append("-")
        
        
    try:
        loc = driver.find_elements_by_xpath("//small[@class = 'ellipsis']")

        for i in loc:
            loc_name = i.text
            Location.append(loc_name)
    
    except NoSuchElementException as e:
        Location.append("-")
    
    
    try:
        tech = driver.find_elements_by_xpath("//div[@class = 'hireSec highlightable']")

        for i in tech:
            tech_name = i.text
            Skills.append(tech_name)
    
    except NoSuchElementException as e:
        Skills.append("-")
```

```python
data()
```
