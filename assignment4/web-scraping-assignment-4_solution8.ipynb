```python
import selenium
import pandas as pd
from selenium import webdriver 
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(r"C://chromedriver.exe")
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
```

```python
Book_name = []
Author_name = []
Volumes_sold = []
Publisher = []
Genre = []
```

```python
bname = driver.find_elements_by_xpath("//table[@class = 'in-article sortable']//tbody//tr//td[2]")

for i in bname :
    bk_name = i.text
    Book_name.append(bk_name)

#Book_name
```

```python
aname = driver.find_elements_by_xpath("//table[@class = 'in-article sortable']//tbody//tr//td[3]")

for i in aname :
    at_name = i.text
    Author_name.append(at_name)

#Author_name
```

```python
volume = driver.find_elements_by_xpath("//table[@class = 'in-article sortable']//tbody//tr//td[4]")

for i in volume :
    vol = i.text
    Volumes_sold.append(vol)

#Volumes_sold
```

```python
pub = driver.find_elements_by_xpath("//table[@class = 'in-article sortable']//tbody//tr//td[5]")

for i in pub :
    publ = i.text
    Publisher.append(publ)

#Publisher
```

```python
gen = driver.find_elements_by_xpath("//table[@class = 'in-article sortable']//tbody//tr//td[6]")

for i in gen :
    genr = i.text
    Genre.append(genr)

#Genre
```

```python
Novels= pd.DataFrame({})
Novels['Book Name']= Book_name
Novels['Author Name']= Author_name
Novels['Volumes Sold']= Volumes_sold
Novels['Publisher']= Publisher
Novels['Genre']= Genre


Novels
```
