```python
import selenium
import pandas as pd
from selenium import webdriver 
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(r"C://chromedriver.exe")
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
Name = []
Year_span = []
Genre = []
Run_time = []
Ratings = []
Votes = []
```

```python
sname = driver.find_elements_by_xpath("//h3[@class = 'lister-item-header']//a")

for i in sname :
    s_name = i.text
    Name.append(s_name)

#len(Name)
```

```python
year = driver.find_elements_by_xpath("//span[@class = 'lister-item-year text-muted unbold']")

for i in year :
    span = i.text
    Year_span.append(span)

#len(Year_span)
```

```python
ser = driver.find_elements_by_xpath("//span[@class = 'genre']")

for i in ser :
    gen = i.text
    Genre.append(gen)

#len(Genre)
```

```python
dur = driver.find_elements_by_xpath("//span[@class = 'runtime']")

for i in dur :
    duration = i.text
    Run_time.append(duration)

#len(Run_time)
```

```python
rate = driver.find_elements_by_xpath("//div[@class = 'ipl-rating-star small']//span[2]")

for i in rate :
    rating = i.text
    Ratings.append(rating)

#len(Ratings)
```

```python
vote = driver.find_elements_by_xpath("//span[@name = 'nv']")

for i in vote :
    voting = i.text
    Votes.append(voting)

#len(Votes)
```

```python
Series= pd.DataFrame({})
Series['Series Name']= Name
Series['Year']= Year_span
Series['Genre']= Genre
Series['Run_time']= Run_time
Series['Ratings']= Ratings
Series['Votes']= Votes

Series
```
