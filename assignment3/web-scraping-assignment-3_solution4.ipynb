```python
import selenium
import pandas as pd
from selenium import webdriver 
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
```

```python
driver = webdriver.Chrome(r"C://chromedriver.exe")
driver.get('https://www.flipkart.com/')
```

```python
cancel_button = driver.find_element_by_xpath("//button[@class ='_2KpZ6l _2doB4z']")
cancel_button.click()
```

```python
search_item = driver.find_element_by_xpath("//input[@class='_3704LK']")
search_item.clear()
search_item.send_keys("pixel 4A")

search_button = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button.click()
```

```python
Brand_Name = []
Smartphone_name = [] 
Colour = []
RAM = []
ROM = []
Primary_Camera = []
Secondary_Camera = []
Display_Size = []
Display_Resolution = [] 
Processor = []
Processor_Cores = [] 
Battery_Capacity = [] 
Price = []
Product_URL = []
```

```python
delay = 10
WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'_1fQZEK')))
```

```python
urls = driver.find_elements_by_xpath("//a[@class = '_1fQZEK']")
for i in urls:
    url = i.get_attribute('href')
    Product_URL.append(url)

#Product_URL    
```

```python
len(Product_URL)
```

```python
for page in Product_URL[0:10]:
    
    driver.get(page)
    time.sleep(3)
    #WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'_2KpZ6l _1FH0tX')))
    
    read_more = driver.find_element_by_xpath("//button[@class = '_2KpZ6l _1FH0tX']")
    read_more.click()
    
    try:
        brand = driver.find_element_by_xpath("//span[@class = 'B_NuCI']").text
        brand = brand.split()[0]
        Brand_Name.append(brand)
    
    except NoSuchElementException as e:
        Brand_Name.append("-")
    
    
    
    try:
        name = driver.find_element_by_xpath("//table[@class = '_14cfVK'][1]//tr[3]//td[2]").text
        Smartphone_name.append(name)
    
    except NoSuchElementException as e:
        Smartphone_nameame.append("-")
    
    
    
    try:
        col = driver.find_element_by_xpath("//table[@class = '_14cfVK'][1]//tr[4]//td[2]").text
        Colour.append(col)
    
    except NoSuchElementException as e:
        Colour.append("-")
        
    
    
    try:
        cost = driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']").text
        Price.append(cost)
    
    except NoSuchElementException as e:
        Price.append("-")
        
    
    data = requests.get(page)
    #page
    soup = BeautifulSoup(data.text, 'html.parser')
    req = soup.find_all('table',class_='_14cfVK')
    
    try:
        for i in req[3]:
            rom = i.find_all('li',class_='_21lJbe')
        
            Roms = rom[0].text
            ROM.append(Roms)
    
    except NoSuchElementException as e:
        ROM.append("-")
        
        
    try:
        for i in req[3]:
            ram = i.find_all('li',class_='_21lJbe')
            
            Rams = ram[1].text
            RAM.append(Rams)
    
    except NoSuchElementException as e:
        RAM.append("-")
        
        
        
    try:    
        for j in req[4]:
            data = j.find_all('li',class_='_21lJbe')
        
            if len(data) > 2:
                pcam = data[1].text
                Primary_Camera.append(pcam)
            
            else:
                pcam = data[0].text
                Primary_Camera.append(pcam)
    
    except NoSuchElementException as e:
        Primary_Camera.append("-")
        
    
    try:
        for j in req[4]:
            data = j.find_all('li',class_='_21lJbe')
            
            if len(data) < 4:
                scam = data[1].text
                Secondary_Camera.append(scam)                
                
            else:
                scam = data[4].text
                Secondary_Camera.append(scam)
     
    except NoSuchElementException as e:
        Secondary_Camera.append("-")
        
    
    try:
        for k in req[1]:
            res = k.find_all('li',class_='_21lJbe')
        
            dsize = res[0].text
            Display_Size.append(dsize)
    
    except NoSuchElementException as e:
        Display_Size.append("-")
                                
      
    try:
        for k in req[1]:
            res = k.find_all('li',class_='_21lJbe')
        
            dres = res[1].text
            Display_Resolution.append(dres)
    
                                
    except NoSuchElementException as e:
        Display_Resolution.append("-")
        
    
                                
    try:
        for l in req[2]:
            proc = l.find_all('li',class_='_21lJbe')
        
            proces = proc[1].text
            Processor.append(proces)
        
    except NoSuchElementException as e:
        Processor.append("-")
                                
         
    try:
        for l in req[2]:
            proc = l.find_all('li',class_='_21lJbe')
                                
            proccore = proc[2].text
            Processor_Cores.append(proccore)
     
    except NoSuchElementException as e:
        Processor_Cores.append("-")
                                
                                
    try:    
        for m in req[9]:
            bat = m.find_all('li',class_='_21lJbe')
        
            battery = bat[0].text
            Battery_Capacity.append(battery)
                                
    
    except NoSuchElementException as e:
        Battery_Capacity.append("-")       
    
      
```

```python
Mobile = pd.DataFrame({

'Brand_Name':Brand_Name[0:10],
'Smartphone_name':Smartphone_name[0:10],
'Colour':Colour[0:10],
'RAM':RAM[0:10],
'ROM':ROM[0:10],
'Primary_Camera':Primary_Camera[0:10],
'Secondary_Camera':Secondary_Camera[0:10],
'Display_Size':Display_Size[0:10],
'Display_Resolution': Display_Resolution[0:10],
'Processor':Processor[0:10],
'Processor_Cores': Processor_Cores[0:10],
'Battery_Capacity': Battery_Capacity[0:10],
'Price':Price[0:10],
'Product_URL':Product_URL[0:10]
    
})
```
