```python
Brand_Name = []
Name_of_Product = [] 
Rating = []
No_of_Ratings = [] 
Price = []
Return_Exchange = []
Expected_Delivery = []
Availability = []
Other_Details = []
#Product_URL = []
```

```python
final_link=[]
link_url = driver.find_elements_by_xpath("//li[@class='a-selected' or @class ='a-normal']//a")[0:3]

for i in link_url:
    links = i.get_attribute('href')
    final_link.append(links)

final_link
```

```python
sub_link = []
for page in final_link:
    sublink_url = driver.find_elements_by_xpath("//a[@class= 'a-link-normal a-text-normal']")
    
    for j in sublink_url:
        sub = j.get_attribute('href')
        sub_link.append(sub)


len(sub_link)    
```

```python
for link in sub_link[0:9]:
    
    driver.get(link)
    time.sleep(3)
    
    brand = driver.find_element_by_id('productTitle').text
    brand_nm = brand.split()[0]
    Brand_Name.append(brand_nm)
    
    Name_of_Product.append(brand)
    
    try:    
        ratings = driver.find_element_by_xpath("//span[@data-hook = 'rating-out-of-text']").text
        Rating.append(ratings)
        
    except NoSuchElementException as e:
        Rating.append("-")
        
    
    try:
        noofrate = driver.find_element_by_xpath("//span[@class='a-size-base a-color-secondary'][1]").text
        nrt = noofrate.split()[0]
        No_of_Ratings.append(nrt)
        
    except NoSuchElementException as e:
        No_of_Ratings.append("-")
        
        
    try:
        cost = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockSalePriceString' or @class='a-size-medium a-color-price priceBlockBuyingPriceString']").text
        Price.append(cost)
        
    except NoSuchElementException as e:
        Price.append("-")
        
        
   # try:
   #     rex = driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal'][2]").text
   #     Return_Exchange.append(rex)
   #     
   # except NoSuchElementException as e:
   #     Return_Exchange.append("-")
    
   
    try:
        date = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b").text
        Expected_Delivery.append(date)
        
    except NoSuchElementException as e:
        Expected_Delivery.append("-")
        
       

    try:
        avail = driver.find_element_by_xpath("//div[@id='availability']//span").text
        Availability.append(avail)
        
    except NoSuchElementException as e:
        Availability.append("-")
    
```

```python
User_Data= pd.DataFrame({
'Brand': Brand_Name[0:9],
'Product Name': Name_of_Product[0:9],
'Rating': Rating[0:9],
'No_of_Ratings': No_of_Ratings[0:9],
'Price': Price[0:9],
'Expected_Delivery': Expected_Delivery[0:9],
'Availability': Availability[0:9],
'Product URL': sub_link[0:9]

})
```
